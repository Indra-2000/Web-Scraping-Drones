```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
from pandas import json_normalize
import requests
import re
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
from bs4 import BeautifulSoup
```

```python
df = pd.read_csv("cleaned10_drones_new.csv")
```

```python
df
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df["price"], bins=20, kde=True)
plt.title("Price Distribution")
plt.show() #Price Distribution
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x=df["Rating"])
plt.title("Rating Count")
plt.show() #Rating distribution
```

```python
plt.figure(figsize=(8,4))
df["Brand"].value_counts().plot(kind="bar")
plt.title("Number of Products per Brand")
plt.ylabel("Count")
plt.show() # Brand frequency
```

```python
plt.figure(figsize=(5,4))
sns.boxplot(y=df["price"])
plt.title("Price Outliers")
plt.show() #Boxplot for price 
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x=df["price"], y=df["Rating"])
plt.title("Price vs Rating")
plt.show() # Bivariate price vs Rating
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x="Brand", y="price", data=df)
plt.xticks(rotation=45)
plt.title("Average Price per Brand")
plt.show() #Brand vs price
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x="Brand", y="Rating", data=df)
plt.xticks(rotation=45)
plt.title("Brand vs Rating")
plt.show()  # Brand vs rating
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df[["price","Rating","no_of_Ratings"]].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show() # multivariate 
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="price", y="Rating", hue="Brand", data=df)
plt.title("Price vs Rating vs Brand")
plt.show()
```
